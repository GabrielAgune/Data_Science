```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
songs = pd.read_csv('spotify_songs.csv')
```

```python
songs.head()
```

```python
sizes2 = songs['playlist_genre'].value_counts()
sizes2
```

```python
sizes = songs['playlist_genre'].unique()
sizes
```

```python
sizes2 = [5507, 5746, 4951, 5155, 5431, 6043]
perc = []
total = sum(sizes2)
for i in range(len(sizes2)):
  perc.append(sizes2[i]/total)

ig, ax = plt.subplots()
ax.pie(perc, labels=sizes, autopct='%1.1f%%')
```
